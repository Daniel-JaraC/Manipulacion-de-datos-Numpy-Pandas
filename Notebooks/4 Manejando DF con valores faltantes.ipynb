# Manejando DF con valores faltantes.

```python
df = pd.read_csv('data/Prueba.csv')
df
```

### Alternativa 1) Llenar todos los espacios vac√≠os con el mismo valor.

```python
df['Col4'].fillna(0)
```

```python
# si no redefines el df no se guarda los cambios y pasa lo que sucede abajo. Para evitarlo hay que redefinir el df
# quedar√≠a de la siguiente manera: df['Col4'] = df['Col4'].fillna(0) 
df['Col4']
```

### Alternativa 2) Llenar los espacios vacios usando las celdas cercanas.

```python
df['Col4'].fillna(method = 'backfill')
```

```python
df['Col4'].fillna(method = 'ffill')
```

```python
graph1 = df['Col4'].fillna(method = 'backfill')
graph2 = df['Col4'].fillna(method = 'ffill')

fig = plt.figure()
plt.plot(graph1, label = 'backfill')
plt.plot(graph2, label = 'ffill')
plt.legend()
plt.show()
```

```python
fig = plt.figure()
plt.plot(graph1, label = 'backfill')
plt.legend()
plt.show()

fig = plt.figure()
plt.plot(graph2, label = 'ffill')
plt.legend()
plt.show()
```

```python
graph1 = df['Col4'].fillna(method = 'backfill')[18:25]
graph2 = df['Col4'].fillna(method = 'ffill')[18:25]

fig = plt.figure()
plt.plot(graph1, label = 'backfill')
plt.plot(graph2, label = 'ffill')
plt.legend()
plt.show()
```

### Alternativa 3) Llenar los espacios con el valor promedio.

```python
y_mean = df['Col4'].mean()
y_mean
```

```python
fig = plt.figure()
# la siguiente instrucci√≥n crea una l√≠nea horizontal en el valor de la media y va desde 0 a 40
plt.hlines(y_mean, 0, 40, linestyles='dashed')
plt.plot(df['Col4'])
plt.show()
```

```python
graph1 = df['Col4'].fillna(y_mean)

fig = plt.figure()
plt.plot(graph1)
plt.hlines(y_mean, 0, 40, linestyles='dashed')
plt.show()
```

# Modifiquemos nuestro df.

```python
print(df['Col1'].mean())
print(df['Col2'].mean())
print(df['Col3'].mean())
print(df['Col4'].mean())
print(df['Col5'].mean())
```

```python
# apply hace que una funci√≥n se aplique a lo largo de un eje  
mod_df = df.apply(lambda x: x.fillna(x.mean()))
mod_df
```

<b>Nota:</b> en este ejemplo utilizamos algo llamado "<b>Funciones Lambda</b>", por si gustas leer un poco m√°s del tema.

```python
mod_df.describe()
```
